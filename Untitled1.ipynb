```python
import pandas as pd
import os
import PIL
import numpy as np
from PIL import Image,ImageOps
```

```python
from os import listdir
from os.path import isfile, join
mypath="data/training_images"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
dataset1=pd.read_csv("data/train_solution_bounding_boxes (1).csv")
```

```python
dataset1
```

```python
present=list(dataset1["image"])
```

```python
final_dataset=pd.DataFrame({"Im_Name":[],"Xmin":[],"Ymin":[],"Xmax":[],"Ymax":[],"Image":[]})
final_dataset
for im in onlyfiles:
    imag=Image.open("data/training_images/"+im)
    gray_image = ImageOps.grayscale(imag)
    im_data=np.array(gray_image.getdata())
    im_data=im_data/255
    if im in present:
        im_name,xmin,ymin,xmax,ymax=dataset1[dataset1["image"]==im].iloc[0]
        final_dataset.loc[len(final_dataset.index)] = [im,xmin,ymin,xmax,ymax,im_data]
    else:
        final_dataset.loc[len(final_dataset.index)] = [im,0,0,0,0,im_data]
        
    
```
